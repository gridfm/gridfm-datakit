```python
import pandas as pd
import os
```

```python
pf_node = pd.read_csv("../../data_contingency/case300_ieee/raw/pf_node.csv")
pf_node_output = pd.read_csv("../../data_contingency/case300_ieee/output/GridFM_v0.1.2_case300_ieee_output_pf_node.csv")
```

```python
# show shape of pf_node_output and pf_node
print(pf_node_output.shape)
print(pf_node.shape)
```

```python
pf_node_output.drop(columns=["Unnamed: 0", "PQ", "PV", "REF"], inplace=True)
```

```python
# add "_pred" to the columns PD, QD, PG, QG, PL, QL, VM, VA to pf_node_output
pf_node_output.columns = [col + "_pred" if col not in ["scenario", "bus", "PQ", "PV", "REF"] else col for col in pf_node_output.columns]
pf_node_output
# add pf_node data to pf_node_output on the columns scenario and bus
```

```python
# add pf_node data to pf_node_output on the columns scenario and bus
pf_node_output = pf_node_output.merge(pf_node, on=["scenario", "bus"], how="left")
pf_node_output

```

```python
# save merged data
os.makedirs("../../data_contingency/case300_ieee/merged", exist_ok=True)
pf_node_output.to_csv("../../data_contingency/case300_ieee/merged/GridFM_v0.1.2_case300_ieee_output_pf_node.csv", index=False)
```
